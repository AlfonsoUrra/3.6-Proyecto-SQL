# Funciones 

```python
# Visualización preliminar de datos

def analizar_data(data):
    print('Valores nulos')
    display(data.isnull().sum())
    print()
    print('Número de columnas y filas')
    display(data.shape)
    print()
    print('Númerode duplicados')
    display(data.duplicated().sum())
    print()
    display(data.head())

    
```
